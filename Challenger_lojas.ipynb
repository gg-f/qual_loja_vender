### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
# Faturamento total da Loja 1
loja = pd.read_csv(url)
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"

loja.head()

total_faturamento = loja['Preço'].sum()
print(f'Faturamento total: R${total_faturamento:,.2f}')
```

```python
# Faturamento total da Loja 2

url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
loja2 = pd.read_csv(url2)
loja2.head()

total_faturamento = loja2['Preço'].sum()
print(f'Faturamento total: R${total_faturamento:,.2f}')
```

```python
# Faturamento total da Loja 3

url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
loja3 = pd.read_csv(url3)
loja3.head()

total_faturamento = loja3['Preço'].sum()
print(f'Faturamento total: R${total_faturamento:,.2f}')
```

```python
# Faturamento total da Loja 4
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"
loja4 = pd.read_csv(url4)
loja4.head()

total_faturamento = loja4['Preço'].sum()
print(f'Faturamento total: R${total_faturamento:,.2f}')
```

```python

# Somar o faturamento de cada loja
total_faturamento1 = loja['Preço'].sum()
total_faturamento2 = loja2['Preço'].sum()
total_faturamento3 = loja3['Preço'].sum()
total_faturamento4 = loja4['Preço'].sum()

# Criando listas para o gráfico
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [total_faturamento1, total_faturamento2, total_faturamento3, total_faturamento4]


# DataFrame e ordenação
df = pd.DataFrame({'Loja': lojas, 'Faturamento': faturamentos}).sort_values(by='Faturamento')

# Gradiente mais agressivo (elevando ao quadrado)
normalized = (df['Faturamento'] / df['Faturamento'].max())**2
colors = plt.cm.Blues(normalized)

# Gráfico
plt.title('Faturamento por Loja', fontsize=14, fontweight='bold')
plt.figure(figsize=(10,6))
plt.barh(df['Loja'], df['Faturamento'], color=colors, edgecolor='gray')
for i, v in enumerate(df['Faturamento']):
    plt.text(v*1.01, i, f'R${v:,.0f}'.replace(',', 'X').replace('.', ',').replace('X', '.'), va='center')


#Deixando o fundo mais limpo
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)


plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.gca().set_axisbelow(True)
plt.xticks([])  # remove valores do eixo X
plt.tight_layout()
plt.show()

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria

```python
import pandas as pd
import matplotlib.pyplot as plt
# Suas colunas de categoria
categoria1 = loja['Categoria do Produto']
categoria2 = loja2['Categoria do Produto']
categoria3 = loja3['Categoria do Produto']
categoria4 = loja4['Categoria do Produto']

lojas = [categoria1, categoria2, categoria3, categoria4]
nomes_lojas = ['Loja1', 'Loja2', 'Loja3', 'Loja4']

for i, cat in enumerate(lojas):
    # Conta quantas vezes cada categoria aparece
    contagem = cat.value_counts(normalize=True) * 100  # porcentagem

    plt.figure(figsize=(6, 6))
    plt.pie(
        contagem.values,
        labels=contagem.index,
        autopct='%1.1f%%',
        startangle=90
    )
    plt.title(f'Distribuição de Categorias - {nomes_lojas[i]}')
    plt.axis('equal')  # deixa o gráfico redondo
    plt.show()

```

```python
#Calculanddo a Média das Avaliações

media_avaliacao = round(loja['Avaliação da compra'].mean(),1)
print(f'Média de Avaliação da loja 1:{media_avaliacao}')

media_avaliacao2 = round(loja2['Avaliação da compra'].mean(),1)
print(f'Média de Avaliação da loja 2: {media_avaliacao2}')

media_avaliacao3 = round(loja3['Avaliação da compra'].mean(),1)
print(f'Meida de Avaliação da loja 3: {media_avaliacao3}')

media_avaliacao4 = round(loja4['Avaliação da compra'].mean(),1)
print(f'Media de Avaliação da loja 4: {media_avaliacao4}')

#Atribundo correspondentes
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
medias = [media_avaliacao, media_avaliacao2, media_avaliacao3, media_avaliacao4]

#Gerando Grafico

plt.plot (lojas, medias, marker='o', color='red')

plt.title('Media de Avalições por loja')
plt.xlabel('lojas')
plt.ylabel('Média')

plt.ylim(0, 5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Lendo os arquivos CSV
loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# Selecionando somente a coluna desejada
loja_produto1 = loja1[['Produto']].dropna(subset=['Produto'])
loja_produto2 = loja2[['Produto']].dropna(subset=['Produto'])
loja_produto3 = loja3[['Produto']].dropna(subset=['Produto'])
loja_produto4 = loja4[['Produto']].dropna(subset=['Produto'])

# Mais e menos vendidos da Loja 1
contagem1 = loja_produto1['Produto'].value_counts()
mais_vendido1 = contagem1.idxmax()
menos_vendido1 = contagem1.idxmin()

# Loja 2
contagem2 = loja_produto2['Produto'].value_counts()
mais_vendido2 = contagem2.idxmax()
menos_vendido2 = contagem2.idxmin()

# Loja 3
contagem3 = loja_produto3['Produto'].value_counts()
mais_vendido3 = contagem3.idxmax()
menos_vendido3 = contagem3.idxmin()

# Loja 4
contagem4 = loja_produto4['Produto'].value_counts()
mais_vendido4 = contagem4.idxmax()
menos_vendido4 = contagem4.idxmin()

# Criando DataFrame com os resultados
dados = {
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Mais Vendido (Qtd)': [contagem1.max(), contagem2.max(), contagem3.max(), contagem4.max()],
    'Menos Vendido (Qtd)': [contagem1.min(), contagem2.min(), contagem3.min(), contagem4.min()]
}

df = pd.DataFrame(dados)

# Exibindo o resumo no console
print(df)

# Criando o gráfico comparativo
plt.figure(figsize=(10,6))
plt.bar(df['Loja'], df['Mais Vendido (Qtd)'], label='Mais Vendido', color='royalblue', alpha=0.8)
plt.bar(df['Loja'], df['Menos Vendido (Qtd)'], label='Menos Vendido', color='lightcoral', alpha=0.8)

plt.title('Comparativo de Produtos Mais e Menos Vendidos por Loja')
plt.xlabel('Lojas')
plt.ylabel('Quantidade de Vendas')
plt.legend()
plt.tight_layout()
plt.show()
```
